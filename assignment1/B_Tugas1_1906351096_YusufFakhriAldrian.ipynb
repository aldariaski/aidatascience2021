[15] Lakukan penanganan terhadap missing values, duplikasi data, dan outliers jika ada!


```python
#Duplikasi data
startup_data_cleaned = startup_data.drop_duplicates()

#Missing values
startup_data_cleaned.fillna(startup_data_cleaned.median)

display(startup_data_cleaned.head)

print("Handling outliers")
#Outliers 
for col in startup_data_cleaned.columns:
  #break #kalau dibawah masih error pk break aja
  startup_col = startup_data_cleaned[col]
  if startup_col.dtype != np.float64:
    continue
  if startup_col.empty:
    continue
  sorted(startup_data_cleaned)
  q1_t, q3_t = np.percentile(startup_col,[25,75])
  iqr = q3_t - q1_t
  lower_bound = q1_t -(1.5 * iqr) 
  upper_bound = q3_t +(1.5 * iqr) 
  print(col + " had " + str(sum(startup_col < lower_bound) + sum(startup_col >  upper_bound)) + " outliers.")
  startup_data_cleaned = startup_data_cleaned[startup_col.between(lower_bound, upper_bound)]
  
print("\n\nMissing values, duplications, outliers have been handled")
startup_data = startup_data_cleaned
```

```python
from numpy.linalg import eig
# define a matrix
df = cancer_reg
df = df.filter(['avgAnnCount', 'avgDeathsPerYear', 'incidenceRate', 'incidenceRate', 'studyPerCap', 'TARGET_deathRate'])

# membagi data menjadi X dan y (kelas)
X = df.iloc[:,0:5].values
y = df.iloc[:,5].values

from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(X)

# Menghitung vektor rata-rata
mean_vec = np.mean(X_std, axis=0)

# Menghitung covariance matrix berdasarkan rumus di atas
cov_mat = (X_std - mean_vec).T.dot((X_std - mean_vec))/(X_std.shape[0]-1)

print('Covariance matrix \n%s' %cov_mat)

print('NumPy covariance matrix: \n%s' %np.cov(X_std.T))

eig_vals, eig_vecs = np.linalg.eig(cov_mat)

print('Eigenvectors \n%s' %eig_vecs) 
print('\nEigenvalues \n%s' %eig_vals)

# Membuat list dari tuple (eigenvalue, eigenvector) 
eig_pairs = [(np. abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]

# Mengurutkan eig_pairs dari nilai eigen tertinggi ke terendah
eig_pairs.sort() 
eig_pairs.reverse() 

print('Eigenvalues in descending order:') 
for i in eig_pairs: 
 print(i[0])

import matplotlib.pyplot as plt

tot = sum(np.abs(eig_vals)) 
var_exp = [(i / tot) for i in sorted(np.abs(eig_vals), reverse=True)]

cum_var_exp = np.cumsum(var_exp)
plt.bar(range(1, eig_vals.size + 1), var_exp, alpha=0.5, align='center', label='Individual') 
plt.step(range(1, eig_vals.size + 1), cum_var_exp, where='mid',  label='Cumulative') 
plt.ylabel('Explained variance ratio') 
plt.xlabel('Principal components') 
plt.legend(loc='best') 
plt.tight_layout() 
plt.savefig('./fig-pca-var-exp.png', dpi=300) 
plt.show()

matrix_w = np.hstack((eig_pairs[0][1].reshape(5,1), eig_pairs[1][1].reshape(5,1))) 
print('Matrix W:\n', matrix_w)

Y = X_std.dot(matrix_w)

#Plot the result 
colors = ['r', 'b', 'g'] 
markers = ['1', '2', '3'] 

for l, c, m in zip(np.unique(y), colors, markers): 
 plt.scatter(Y[y==l, 0], 
 Y[y==l, 1], 
 c=c, label=l, marker=m)
 
plt.title('Y') 
plt.xlabel('PC 1') 
plt.ylabel('PC 2') 
plt.legend(loc='lower left') 
plt.tight_layout() 
plt.savefig('./fig-pca-z.png', dpi=300) 
plt.show()
```
